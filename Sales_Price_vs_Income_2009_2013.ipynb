```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
#Extracting URL information for use in our datasets
url = 'https://en.wikipedia.org/wiki/List_of_United_States_counties_by_per_capita_income'
```

```python
tables = pd.read_html(url)
tables
```

```python
type(tables)
```

```python
df = tables[2]
df.columns = ['Rank', 'County or county-equivalent State', 'federal district or territory','Per Capitaincome', 'Medianhouseholdincome', 'Medianfamilyincome', 'Population', 'Number ofhouseholds ']
df.head()
```

```python
new_df = df[['federal district or territory', 'Medianhouseholdincome','Medianfamilyincome']].copy()
new_df['Medianhouseholdincome'] = new_df['Medianhouseholdincome'].str.replace('$','')
new_df['Medianhouseholdincome'] = new_df['Medianhouseholdincome'].str.replace(',','')
new_df['Medianfamilyincome'] = new_df['Medianfamilyincome'].str.replace('$','')
new_df['Medianfamilyincome'] = new_df['Medianfamilyincome'].str.replace(',','')
new_df['Medianfamilyincome'] = new_df['Medianfamilyincome'].astype(float)
new_df['Medianhouseholdincome'] = new_df['Medianhouseholdincome'].astype(float)
#new_df['Medianhouseholdincome'] = pd.to_numeric(new_df['Medianhouseholdincome'])
new_df = new_df.rename(columns={'federal district or territory':'state', 'Medianhouseholdincome':'median_household_income','Medianfamilyincome':'median_family_income'})
income_df = new_df.groupby(['state']).mean().round(2)
income_df
```

```python
income_df = income_df.drop(['American Samoa', 'District of Columbia', 'Guam', 'Northern Mariana Islands', 'Puerto Rico', 'U.S. Virgin Islands', 'United States','Vermont','Alaska'])
income_df.reset_index(level = 0, inplace = True)
```

```python
income_df
```

```python
income_df['state'] = income_df['state'].astype(str)
income_df
```

```python
#Extracting CSV File and Transforming it for use in our Database

csv = "../ETL Project/Resources/Sale_Prices_State.csv"
sales_df = pd.read_csv(csv)

sales_df = sales_df.fillna(0)
sales_df = sales_df.sort_values(by = ['RegionName'],ascending = True)
sales_df = sales_df.reset_index()

```

```python
#transform dataset to only information that we require
new2_sales_df = new_sales_df[['RegionName', 'Average Sales 09-13']]
new2_sales_df = new2_sales_df.reset_index(drop= True)
new2_sales_df = new2_sales_df.drop(7)
new2_sales_df = new2_sales_df.reset_index(drop = True)
```
